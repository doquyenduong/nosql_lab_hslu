### Passenger frequencies at SBB stations and stops
Passenger frequencies at SBB stations and stops according to ‘Ein-Aussteiger Datenbank’
Last update Sept. 2019 with 2018 figures.
The figures include the number of boarding and alighting passengers for the following transport companies:
BLS, MGB, MOB, Oensingen-Balsthal-Bahn, RegionAlps, RhB, SBB, SBB GmbH, SOB, SZU, Thurbo, TILO, TRAVYS, Zentralbahn.

For larger stations, the comments state which rail transport companies are not included in the figures for which stations.

The list comprises all stations belonging to MGB, RhB, SBB, SOB, Thurbo or Zentralbahn that are operated to a relevant extent by these transport companies or one of their subsidiaries.
Explanation of designations
Stop abbreviation = Didok code for the station/stop (Didok: service documentation).
Station_Stop = Name of the station/stop.
Canton = What canton the station/stop belongs to.
Owner = Owner of the station (infrastructure manager).

* ADT = dtv = Average daily traffic (Monday to Sunday) in 2018. 
* AWT = dwt = Average weekday traffic volume (Monday to Friday) in 2018. 
* ANWT = dnwv = Average non-weekday traffic volume (Saturdays, Sunday, public holidays) in 2018
Comments for additional information on the entry.
https://data.sbb.ch/explore/dataset/passagierfrequenz/information/

```python
import pymongo
from pprint import pprint
import pandas as pd
import requests
import json
from bson.regex import Regex
from collections import OrderedDict
```

```python
# pandas configuration
pd.set_option('display.precision', 2)
pd.set_option('display.max_rows', 30)
pd.set_option('display.max_colwidth', 50)
# pd.describe_option('max_rows')
# pd.describe_option('precision')
# pd.describe_option('max_colwidth')
```

```python
# API and Database details, here multiple API if u want
API_URL = "https://data.sbb.ch/api/records/1.0/search/?dataset=passagierfrequenz&q=&rows=906&facet=code&facet=bezugsjahr&facet=eigner&facet=dtv&facet=dwv&facet=dnwv"
CNX_STR = "localhost:27017"
DB_NAME = "db_train_station"
COLL_NAME = "passenger_frequencies"
```

```python
# connection to MongoDB
client = pymongo.MongoClient(CNX_STR)
db = client[DB_NAME]
passenger_frequencies = db[COLL_NAME]
```

## ETL

### Remove all existing documents -> Reset collection

```python
passenger_frequencies.drop()
passenger_frequencies.count_documents({})
```

###  Fetch data

```python
# fetch JSON from API_URL
r = requests.get(API_URL)
data = json.loads(r.text)
```

```python
# here is to show ur data
print(r.text[0:500])
```

```python
data["records"]
```

### Insert into MongoDB

```python
# insert the list of jokes in "value" into MongoDB collection "jokes"
passenger_frequencies.insert_many(data['records']);
```

```python
# count number of documents inserted
passenger_frequencies.count_documents({})
```

```python
c = passenger_frequencies.aggregate([
      {"$limit": 1},
])

for doc in c:
     pprint(f"{doc}"[:500])
```

### Transform

```python
# assign id to _id and remove id
c = passenger_frequencies.aggregate([
    {"$project": {"_id": "$id", "joke": 1, "categories": 1}},
])
pd.DataFrame(c)
```

```python
# Check for empty value
print("1.", "~"*30)
# find document where field value is None
for d in passenger_frequencies.fields.find({"bahnhof_haltestelle": None}):
  pprint(d)
print()

print("2.", "~"*30)
# find document where field value is type 10 == None
for d in passenger_frequencies.fields.find({"bahnhof_haltestelle": {"$type": 10}}):
  pprint(d)
print()

print("3.", "~"*30)
# find document where filed name is missing
for d in passenger_frequencies.fields.find({"bahnhof_haltestelle": {"$exists": False}}):
  pprint(d)
print()
```
