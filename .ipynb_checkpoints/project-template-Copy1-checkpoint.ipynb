###  Fetch data

```python
#url for api
url = "https://api.teleport.org/api/urban_areas/"
r = requests.get(url)

data = json.loads(r.text)
data
```

```python
# Get only relevant documents
href = data["_links"]["ua:item"]
href
```

```python
# Change dictionary to dataframe to check easier
df_href = pd.DataFrame(href)
df_href
```

```python
# Define the string to add at the end of the api
str1 = "scores/"

# Change href from object to string and add the string at the end
df_href['new_href'] = df_href['href'].astype("string") + str1
df_href['new_href']
```

```python
# Make a for loop through 266 urls to get urban scores of 266 cities 
# list_new = []

for index, row in df_href.iterrows():
    # loop through the url in the column new_href & still keep urban area name
    url = df_href["new_href"][index]
    r1 = requests.get(url)
    data1 = json.loads(r1.text)
    time.sleep(1)    
#     list_new.append(data1)
    df_href.at[index,'new_thing'] = data1

df_href
```

```python
# Change dataframe to dictionary, get only 2 columns we need
dict_area_score = dict(zip(df_href.name, df_href.new_thing))
dict_area_score
```

```python
# Write the dictionary into a new json file
json.dump(dict_area_score, open("dict_area_score.json", 'w'))
```
