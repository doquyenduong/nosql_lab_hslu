##  Requirements & Configuration

```python
! pip3 list | findstr "pymongo dnspython pandas"
```

```python
import pymongo
from pprint import pprint
import pandas as pd
import requests
import json
```

```python
# pandas configuration
pd.set_option('precision', 2)
pd.set_option('max_rows', 30)
pd.set_option('max_colwidth', 50)
# pd.describe_option('max_rows')
# pd.describe_option('precision')
# pd.describe_option('max_colwidth')
```

```python
# API and Database details, here multiple API if u want
API_URL = "https://api.icndb.com/jokes/?exclude=[explicit]"
CNX_STR = "localhost:27017"
DB_NAME = "icndb"
COLL_NAME = "jokes"
```

```python
# connection to MongoDB
client = pymongo.MongoClient(CNX_STR)
db = client[DB_NAME]
jokes = db[COLL_NAME]
```

## ETL

### Remove all existing documents -> Reset collection

```python
jokes.drop()
jokes.count_documents({})
```

###  Fetch data

```python
# fetch JSON from API_URL
r = requests.get(API_URL)
data = json.loads(r.text)
```

```python
# here is to show ur data
print(r.text[0:500])
```

```python
data["value"]
```

### Insert into MongoDB

```python
# insert the list of jokes in "value" into MongoDB collection "jokes"
jokes.insert_many(data['value']);
```

```python
# count number of documents inserted
jokes.count_documents({})
```

```python
c = jokes.aggregate([
      {"$limit": 1},
])

for doc in c:
     pprint(f"{doc}"[:500])
```

```python
c = jokes.aggregate([
      {"$limit": 5},
])

pd.DataFrame(c)
```

### Transform

```python
# assign id to _id and remove id
c = jokes.aggregate([
    {"$project": {"_id": "$id", "joke": 1, "categories": 1}},
])
pd.DataFrame(c)
```

### Categories

```python
c = jokes.aggregate([
    {"$project": {"joke": 0}},
    {"$unwind": "$categories"},
    {"$group": {"_id": "$categories", "count": {"$sum": 1}}},
 ])

pd.DataFrame(c)
```

```python
c = jokes.aggregate([
    {"$match": {"categories":  {"$in" : ["nerdy"]}}},
])

pd.DataFrame(c)
```
